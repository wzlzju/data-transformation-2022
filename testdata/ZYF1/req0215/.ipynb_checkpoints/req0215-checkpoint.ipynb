##### Iris

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
iris_df.head()
```

```python
# Normalization so thatåŽç»­ç”»å›¾æ—¶, å›¾ä¸Šçš„æ•°æ®èƒ½å’Œtargetç›¸æ¯”è¾ƒ(targetä¸ºèŠ±çš„å“ç§)
from sklearn.preprocessing import MinMaxScaler
norm = MinMaxScaler().fit(iris_df)
iris_df_norm = norm.transform(iris_df)
```

The formula for normalization:
$$X' = \frac{X - X_{min}}{X_{max} - X_{min}}$$

```python
iris_normed_df = pd.DataFrame(iris_df_norm, columns = iris_df.columns)
iris_normed_df.head()
```

```python
# matplotlib.style.use("seaborn")
iris_normed_df[["sepal length (cm)", "sepal width (cm)", "target"]].plot()
```

```python
iris_normed_df[["petal length (cm)", "petal width (cm)", "target"]].plot()
```

##### COVID-19 Weekly Trends

```python
covid = pd.read_csv("covid.csv")
covid.head()

# ref: https://www.kaggle.com/anandhuh/covid19-weekly-trends-in-world-latest-data
```

```python
covid.shape
```

```python
# select out columns and caculate proportions
percs = covid.copy()
percs["Cases Increase % (to Population)"] = (percs["Cases in the last 7 days"] - percs["Cases in the preceding 7 days"]) / percs["Population"]
percs["Deaths Increase % (to Population)"] = (percs["Deaths in the last 7 days"] - percs["Deaths in the preceding 7 days"]) / percs["Population"]
percs.head()
```

```python
pd.options.display.float_format = '{:.10f}'.format
prop = percs[["Country, Other", "Cases Increase % (to Population)", "Deaths Increase % (to Population)"]]
prop = prop.set_index("Country, Other")
prop.head()
```

```python
# rescaling the data using normalization for better visualization
norm = MinMaxScaler().fit(prop)
prop_norm = norm.transform(prop)
```

```python
prop_normed_df_cols = [c + "(rescaled)" for c in prop.columns]
prop_normed_df = pd.DataFrame(prop_norm, columns = prop_normed_df_cols)
# prop_normed_df["Population"] = percs["Population"]
prop_normed_df = prop_normed_df.T
prop_normed_df.columns = prop.index
prop_normed_df.head()
```

```python
# To avoid overplotting, only select top5 countries ranked by population(later unused as found not promising)
popu5 = percs.sort_values("Population", ascending = False)
popu5_list = popu5.iloc[:5, 0].to_list()
popu5_list
```

```python
# calculate ratio
prop_final = prop_normed_df[popu5_list].T.reset_index()
prop_final["Ratio (Deaths/Cases)"] = prop_final["Deaths Increase % (to Population)(rescaled)"] / prop_final["Cases Increase % (to Population)(rescaled)"]
percs_final = prop_final.iloc[:, :3].melt(id_vars = "Country, Other", var_name = "Type of Increase", value_name = "Rates")
ratio_final = prop_final[["Country, Other", "Ratio (Deaths/Cases)"]]
percs_final
```

```python
ratio_final
```

```python
sns.set(style = "whitegrid")
fig, ax1 = plt.subplots(figsize = (8, 8))
sns.barplot(data = percs_final, x = "Country, Other", y = "Rates", hue = "Type of Increase")
ax1.set_ylabel("Percentages")
plt.legend(bbox_to_anchor = (1.1, 1), loc = 2, borderaxespad = 0.)
ax2 = ax1.twinx()
sns.lineplot(data = ratio_final, 
             palette = "Accent",
             markers = "o"
            )
ax2.set_ylabel("Ratio")
plt.legend(bbox_to_anchor = (1.1, 0.9), loc = 2, borderaxespad = 0.);
# plt.xticks(rotation=90);
```
